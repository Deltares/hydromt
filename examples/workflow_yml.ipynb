# Workflow YAML

You can build models using a declarative workflow `YAML` file. This yaml specifies what model you are building, which components are in that model and which steps are included.

In this case we are building a generic model with only core components, a configuration component: `ConfigComponent`, and a `GridComponent`

Let's take a look at the model config file:

```python
%cat ../tests/data/test_model_config.yml
```

In this case we use the `hydromt build` command line interface (CLI) of HydroMT. First we inquire how to use the build command:

```python
! hydromt build --help
```

In the current environment, only one model is available:

```python
! hydromt --models
```

Let's look into the options that we specify in the next block:
- We specify our `workflow.yml` file using the `-i` option.
- We specify our data catalogs using the `-d` option. The data requested for this area (see `grid.create_from_region`) is in the artifact data shipped together with HydroMT.
The vito_reclass data is part of the examples folder, and a small extra catalog is present there, which explains the two data catalogs that we add.
- We want more logging information than default (warning level), so we specify the `-v` option.
- If there is already a model present, we override the model using the `--fo` options
- We then specify the model that we are building (just `model`), and where the model will end up (`yaml_config_model`)

```python
! hydromt build -i ../tests/data/test_model_config.yml -d ../data/catalogs/artifact_data/v1.0.0/data_catalog.yml -d data/vito_reclass.yml -v --fo model ./yaml_config_model
```

Let's take a look at the output in the `yaml_config_model` directory

```python
%ls yaml_config_model
```

Here we can see the config we specified, the hydromt log file for this model, and the `grid` model we built.

```python
%ls yaml_config_model/grid
```

The grid model has a region geojson file, as well as a grid netcdf file
